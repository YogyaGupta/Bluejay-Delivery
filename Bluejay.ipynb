```python
import pandas as pd
from datetime import datetime, timedelta
file_path = '/content/drive/MyDrive/Employee.csv'
df = pd.read_csv(file_path)
df = df.dropna()
df
```

```python
df['Formatted'] = pd.to_datetime(df['Time'])
df['date'] = df['Formatted'].dt.date
df['time'] = df['Formatted'].dt.time
df['formated out'] = pd.to_datetime(df['Time Out'])
df['time out'] = df['formated out'].dt.time
df['date out'] = df['formated out'].dt.date
df.head(10)
```

```python
#Converting the type of the date and time
df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')
df['time out'] = pd.to_datetime(df['time out'], format='%H:%M:%S')
# Calculate the time difference between consecutive rows for each employee
df['TimeDiff'] = df['time out'] - df['time']
```

```python
# a) Employees who have worked for 7 consecutive days
seven_consecutive_days = df[df['TimeDiff'] <= pd.Timedelta(days=1)].groupby('Employee Name').filter(lambda x: len(x) >= 7)
print("Employees who have worked for 7 consecutive days:")
seven_consecutive_days[['Employee Name', 'Position Status','Position ID']].drop_duplicates()
```

```python
# b) Employees who have less than 10 hours of time between shifts but greater than 1 hour
time_between_shifts = df[(df['TimeDiff'] >= pd.Timedelta(hours=1)) & (df['TimeDiff'] < pd.Timedelta(hours=10))]
print("Employees who have less than 10 hours between shifts but greater than 1 hour:")
time_between_shifts[['Employee Name', 'Position Status','Position ID']].drop_duplicates()
```

```python
# c) Employees who have worked for more than 14 hours in a single shift
more_than_14_hours = df[df['TimeDiff'] >= pd.Timedelta(hours=14)]
print("Employees who have worked for more than 14 hours in a single shift:")
more_than_14_hours[['Employee Name', 'Position Status','Position ID']].drop_duplicates()
```
